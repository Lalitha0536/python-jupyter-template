### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

# Example input
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Input Validation
try:
    if len(x) != len(y):
        raise ValueError("x and y arrays must be of the same length.")
except Exception as e:
    print("Error:", e)
else:
    plt.figure(figsize=(8, 5))
    plt.scatter(x, y, color='purple')
    for i in range(len(x)):
        plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(0,5), ha='center')
    plt.title("Scatter Plot with Annotations")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.show()
```
