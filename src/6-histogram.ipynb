### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Error Handling and Validation
try:
    if not all(isinstance(i, (int, float)) for i in data):
        raise TypeError("All histogram values must be numeric.")

    if any(i < 0 for i in data):
        raise ValueError("Histogram values should be non-negative.")

    if len(data) < 2:
        raise ValueError("Histogram needs at least 2 data points.")

    # Optional: detect extreme outliers
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    upper_limit = q3 + 1.5 * iqr
    lower_limit = q1 - 1.5 * iqr
    outliers = [i for i in data if i < lower_limit or i > upper_limit]
    if outliers:
        print(f"Warning: Detected possible outliers: {outliers}")

except Exception as e:
    print("Error:", e)

else:
    plt.figure(figsize=(8, 5))
    plt.hist(data, bins=7, color='skyblue', edgecolor='black')
    plt.title("Distribution of Values")
    plt.xlabel("Value Range")
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```
