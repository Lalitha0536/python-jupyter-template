### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Subplot data
x = list(range(0, 10))
y = [i**2 for i in x]
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 5]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['X', 'Y', 'Z']
pie_sizes = [30, 50, 30]

# Error Handling
try:
    assert len(x) == len(y), "x and y must be same length for line plot"
    assert len(categories) == len(values), "categories and values must match"
    assert len(scatter_x) == len(scatter_y), "scatter x and y must match"
    assert len(pie_labels) == len(pie_sizes), "pie labels and sizes must match"
    if any(s <= 0 for s in pie_sizes):
        raise ValueError("Pie sizes must be positive")
except Exception as e:
    print("Error:", e)
else:
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))

    # Line Plot
    axs[0, 0].plot(x, y, marker='o')
    axs[0, 0].set_title("Line Plot")
    axs[0, 0].set_xlabel("x")
    axs[0, 0].set_ylabel("y = x^2")
    
    # Bar Plot
    axs[0, 1].bar(categories, values, color='skyblue')
    axs[0, 1].set_title("Bar Plot")
    
    # Scatter Plot
    axs[1, 0].scatter(scatter_x, scatter_y, color='green')
    axs[1, 0].set_title("Scatter Plot")
    
    # Pie Chart
    axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
    axs[1, 1].set_title("Pie Chart")
    axs[1, 1].axis('equal')

    plt.tight_layout()
    plt.show()
```
